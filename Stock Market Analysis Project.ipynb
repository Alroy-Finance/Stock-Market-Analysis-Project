```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Select stocks
stocks = ['AAPL', 'TSLA', 'MSFT', 'GOOGL']

# Download data (last 1 year)
data = yf.download(stocks, start='2024-01-01', end='2026-01-01',auto_adjust = False)['Adj Close']

# Show first few rows
print(data.head())

# 1. Calculate Daily Returns
returns = data.pct_change()


# 2. Calculate Risk & Return
mean_returns = returns.mean()
risk = returns.std()

print("\nAverage Returns:\n", mean_returns)
print("\nRisk (Volatility):\n", risk)

# 3. Plot Stock Prices
data.plot(figsize=(10,5))
plt.title("Stock Price Trends")
plt.show()

# 4. Moving Averages (AAPL example)
aapl = data['AAPL']
ma50 = aapl.rolling(window=50).mean()
ma200 = aapl.rolling(window=200).mean()

plt.figure(figsize=(10,5))
plt.plot(aapl, label='AAPL Price')
plt.plot(ma50, label='50-Day MA')
plt.plot(ma200, label='200-Day MA')
plt.legend()
plt.title("AAPL Moving Averages")
plt.show()

# 5. Moving Averages (TSLA example)
tsla = data['TSLA']
ma50_1 = tsla.rolling(window=50).mean()
ma200_1 = tsla.rolling(window=200).mean()

plt.figure(figsize=(10,5))
plt.plot(tsla, label='AAPL Price')
plt.plot(ma50_1, label='50-Day MA')
plt.plot(ma200_1, label='200-Day MA')
plt.legend()
plt.title("TSLA Moving Averages")
plt.show()
```

**Conclusions**

**Tech stocks showed consistent upward trends over the period  
 Tesla exhibited higher volatility compared to Microsoft and Apple  
 Moving averages indicated long-term bullish trends for AAPL**
